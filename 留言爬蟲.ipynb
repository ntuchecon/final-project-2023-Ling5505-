```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

elements = WebDriverWait(driver, 20).until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, "GHT2ce")))
result_list_2 = []

# 下面的迴圈有點障礙，除錯會將沒有留言的內容也刪掉@@，得不到足額的rows
# 我經過幾次查證發現，迴圈在印出文字評論後會多伴隨一次no_text，
# 造成原始的df_comment會有兩倍的rows，因次我在最後再篩出希望取得的部分=df_comment_filtered

for element in elements:
    try:
        # 在每個 <div class="GHT2ce"> 元素下尋找 <div class="MyEned">
        myened_element = element.find_element(By.CLASS_NAME, "MyEned")  # 使用 timeout 避免等待太長時間
        
        # 繼續尋找 <span class="wiI7pd">
        span_element = myened_element.find_element(By.CLASS_NAME, "wiI7pd")  # 使用 timeout 避免等待太長時間
        text_content = span_element.text

        # 移除 "\n" 和標點符號
        cleaned_text = ''.join(char for char in text_content if char != '\n')
        result_list_2.append(cleaned_text)
    except:
        # 如果找不到 "wiI7pd" 元素，直接跳過這個 GHT2ce 元素
        result_list_2.append("no_text")

df_comment = pd.DataFrame(result_list_2, columns=["評論"])
# 選取奇數行
df_comment_filtered = df_comment[df_comment.index % 2 != 0]
# 重新設定索引
df_comment_filtered.reset_index(drop=True, inplace=True)
# 檢查結果
# print(df_comment_filtered.tail(30))
print(df_comment_filtered)

```
